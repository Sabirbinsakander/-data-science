```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.w3schools.com/python/default.asp'
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
title = soup.title.string
print(f"Title: {title}")
```

```python
paragraphs = soup.find_all("p")
```

```python
for i, paragraph in enumerate(paragraphs, start=1):
    print(f"Paragraph {i}: {paragraph.text}")
```

```python
if paragraphs:
    first_paragraph = paragraphs[0].text
    word_count = len(first_paragraph.split())
    print(f"\nWord count in first paragraph: {word_count}")
else:
    print("\nNo paragraphs found on the page.")
```

```python
import requests
from bs4 import BeautifulSoup

# URL for W3Schools Python tutorial page
url = "https://www.w3schools.com/python/default.asp"

response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

# Extract and print the page title
if soup.title:
    title = soup.title.string
else:
    title = "Title not found"
print(f"Page Title: {title}\n")

# Find and print all main headings (h2 tags)
headings = soup.find_all("")
if headings:
    print("Main Headings:")
    for i, heading in enumerate(headings, start=1):
        print(f"  {i}. {heading.text.strip()}")
else:
    print("No main headings (h2) found on the page.")

# You can extend this to extract other elements like paragraphs, links, etc.
# For example, to get paragraphs:
# paragraphs = soup.find_all("p")
# if paragraphs:
#     print("\nFirst 3 Paragraphs:")
#     for i, paragraph in enumerate(paragraphs[:3], start=1):
#         print(f"  {i}. {paragraph.text.strip()}")
```

```python
import requests
from bs4 import BeautifulSoup


url = "https://www.w3schools.com/python/default.asp"


response = requests.get(url)


soup = BeautifulSoup(response.text, 'html.parser')




title = soup.title.string
print(f"Title: {title}")




paragraphs = soup.find_all("p")



for i, paragraph in enumerate(paragraphs, start=1):
    print(f"Paragraph {i}: {paragraph.text}")





if paragraphs:
    first_paragraph = paragraphs[0].text
    word_count = len(first_paragraph.split())
    print(f"\nWord count in first paragraph: {word_count}")
else:
    print("\nNo paragraphs found on the page.")

```
