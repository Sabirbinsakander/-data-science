```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

df = pd.read_excel('/content/sample_data/Online Retail.xlsx')
df.head()
```

```python
df.Country.unique()
```

```python
#Stripping extra spaces in the discription
df['Description'] = df['Description'].str.strip()

#Dropping the rows without any invoice number
df.dropna(axis = 0, subset =['InvoiceNo'], inplace = True)
df['InvoiceNo'] = df['InvoiceNo'].astype('str')

#Dropping all transactions which were done on credit
df = df[~df['InvoiceNo'].str.contains('C')]
```

```python
basket_Italy = (df[df['Country'] == "Italy"]
            .groupby(['InvoiceNo', 'Description'])['Quantity']
            .sum().unstack().reset_index().fillna(0)
            .set_index('InvoiceNo'))
basket_Italy
```

```python
def hot_encode(x):
  if x <= 0:
      return 0
  if x >= 1:
      return 1

basket_encoded = basket_Italy.applymap(hot_encode)
basket_Italy = basket_encoded
basket_Italy
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

frq_items = apriori(basket_Italy, min_support = 0.05, use_colnames = True)

#Collecting the inferred rules in a dataframe

num_itemsets = len(frq_items)
rules = association_rules(frq_items, metric = "lift", min_threshold = 1, num_itemsets = num_itemsets)
rules = rules.sort_values(['confidence', 'lift'], ascending = [False, False])
rules
```
