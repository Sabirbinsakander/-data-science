## **Unwanted Value Handaling**

```python
cnt = 0
for row in df['Food Item']:
    try:
        int(row)
        df.loc[cnt, 'Food Item'] = np.nan
    except ValueError:
      pass
    cnt += 1

df
```

```python
df.loc[15, 'Food Item'] = 'Burger'
df
```

## **Fill mode value to NaN value**

```python
df ['Food Item'].fillna(df['Food Item'].mode()[0], inplace=True)
```

```python
df
```

```python
df ['Age'] = df ['Age'].fillna(df['Age'].mean())
```

```python
df
```

### **Median value**

```python
df ['Age'] = df ['Age'].fillna(df.groupby('Food Item')['Age'].transform('median'))
```

```python
df
```

## **Bfill & ffill**

```python
df['Price']=df['Price'].fillna(method='bfill')
```

```python
df
```

```python
df ['Working Status'] = df ['Working Status'].fillna(method='ffill')
```

```python
df
```

### Type Converstion

```python
df .Age=df.Age.astype('int64')
```

### **Level Endcoding**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Food Item'] = le.fit_transform(df['Food Item'])
df
```

### **Mapping Function**

```python
mapping = {'Yes':1, 'No':0}
df['Working Status'] = df['Working Status'].map(mapping)
df
```

### **Scaling **

### **Nomalization**

```python
from sklearn.preprocessing import MinMaxScaler
# Identify Numeric and Non-Numeric Column
numeric_columns = df.select_dtypes(include=['number']).columns
non_numeric_columns = df.select_dtypes(exclude=['number']).columns
#Initialize the scaler
scaler = MinMaxScaler()
#Fit and Transfrom only The numeric Column
scaled_numeric = scaler.fit_transform(df[numeric_columns])
scaled_numeric_df = pd.DataFrame(scaled_numeric, columns=numeric_columns)
#Combine Scaled Numeric Columns with non-numeric columns
scaled_df = pd.concat([df[non_numeric_columns].reset_index(drop=True), scaled_numeric_df.reset_index(drop=True)], axis = 1)
scaled_df
```

### **Standardization**

```python
from sklearn.preprocessing import StandardScaler
```

```python
#Indentify numeric and non-numeric columns
numeric_columns = df.select_dtypes(include=[np.number]).columns
non_numeric_columns = df.select_dtypes(exclude=[np.number]).columns

#Initialize the scaler
ss = StandardScaler()

#Fit and transform only the numeric columns
scaled_numeric = ss.fit_transform(df[numeric_columns])
#convert the scaled numeric data back to DataFrame
scaled_numeric_df = pd.DataFrame(scaled_numeric, columns=numeric_columns)

#Concatenate the scaled numeric columns with the non numeric colums
scaled_df = pd.concat([df[non_numeric_columns].reset_index(drop=True), scaled_numeric_df.reset_index(drop=True)], axis=1)
print(scaled_df)
```

### **Robust Scaler**

```python
from sklearn.preprocessing import RobustScaler
#Indentify numeric and non-numeric columns
numeric_columns = df.select_dtypes(include=[np.number]).columns
non_numeric_columns = df.select_dtypes(exclude=[np.number]).columns
#Initialize this scaler
robust_scaler = RobustScaler()
#Fit and transform only the numeric columns
scaled_numeric = robust_scaler.fit_transform(df[numeric_columns])
#Convert the scaled numeric data back to dataframe
scaled_numeric_df = pd.DataFrame(scaled_numeric, columns=numeric_columns)
#Concatenate the scaled numeric columns with the non numeric columns
scaled_df = pd.concat([df[non_numeric_columns].reset_index(drop=True), scaled_numeric_df.reset_index(drop=True)], axis=1)
#Display the scaled Dataframe
scaled_df.head()
```
