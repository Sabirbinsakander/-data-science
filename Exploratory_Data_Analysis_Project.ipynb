```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
url = 'https://raw.githubusercontent.com/Rubaiya-Razin/Data-Science-Course/refs/heads/main/iris.csv'
df = pd.read_csv(url)
```

```python
#Display the first 5 rows
df.head()
```

```python
df.tail()
```

```python
#Check Null Value
df.isnull().values.any()
```

```python
df.dtypes
```

```python
#Describe dataset
df.describe()
```

```python
#Select Specific Column
selected_column = df[['sepal_length','sepal_width', 'petal_length', 'petal_width']]
#Calculet correlation matrix
correlation_matrix = selected_column.corr()
correlation_matrix
```

```python
#Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu', linewidths=0.7)
plt.title('Correlation Heatmap of selected Feautures')
plt.show()
```

### **Box Plot**

```python
sns.boxplot(x='species', y='sepal_length', data=df)
```

### **Histogram**

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Set the aesthetic style of the plot
sns.set(style='whitegrid')

# Histogram for the numerical feautures
df.hist(figsize=(10, 8), bins=20, color='skyblue')
plt.suptitle('Feature Distributions ', fontsize=16)

plt.show()
#
```

### **Scatter Plot**

```python
#Create a scatter plot between 'Sepal_length' and 'sepal_width'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df, palette='Set2')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
#
```

### **Pair plot**

```python
#Pair plot to visualize relationships between numerical feautures and specics
sns.pairplot(df, hue='species', palette='Set2')
plt.suptitle('Pair Plot of Numerical Features', y=1.02)
plt.show()
```
